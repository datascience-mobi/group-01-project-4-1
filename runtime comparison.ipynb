```python
import time
import warnings

import numpy as np
import matplotlib.pyplot as plt

from sklearn import cluster, datasets, mixture
from sklearn.neighbors import kneighbors_graph
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

from itertools import cycle, islice

import timeit
```

```python
# distance calculator

def dist(a, b, ax=1):
    return np.linalg.norm(a - b, axis=ax)

# Fixing random state for reproducibility
#np.random.seed(19680801)

```

```python
n_samples=1000
```

```python
blobs = datasets.make_blobs(n_samples=n_samples, random_state=15)

points_blobs=blobs[0]

f1=points_blobs[0:((n_samples)), 0]
f2=points_blobs[0:((n_samples)), 1]

plt.scatter(f1, f2, c='black', s=7)

X = np.array(list(zip(f1, f2)))
```

```python
times_to_repeat = 10

sum_runtime = 0

for i in range (times_to_repeat):
    

    start_time = timeit.default_timer()


    k=2
   
    cx = np.random.randint(f1.min(),f1.max(),size=k)
    cy = np.random.randint(f2.min(),f2.max(),size=k)
    
    C = np.array(list(zip(cx, cy)), dtype=np.float32)

    clusters = np.zeros(len(X))
    C_old = np.zeros(C.shape)
    error = dist(C, C_old, None) # Konvergenz
    from copy import deepcopy

    while error != 0: # 
        for i in range(len(X)):
                distances = dist(C,[X[i]])
                cluster = np.argmin(distances)
                clusters[i] = cluster
        C_old = deepcopy(C)
        for i in range(k):
            points = [X[j] for j in range(len(X)) if clusters[j] == i]
            C[i] = np.mean(points, axis=0)
        error = dist(C, C_old, None)
        

    runtime= timeit.default_timer() - start_time
    
    sum_runtime= sum_runtime+runtime
    
    own_k2= sum_runtime/times_to_repeat
    
print(own_k2)


plt.scatter(f1,f2,c='black',s=7)
plt.scatter(cx,cy,marker='*',c='red',s=100)
    
    
```
